```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

names =[]
urls=[]

def dogs(url):
    
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')

##### find all the h3 tags with class and extract text from them ######

    for i in soup.findAll('h3', class_='breed-type-card__title'):
        ttl=i.getText()
        names.append(ttl)

##### find all the div tags with class and extract href from them ######
        
    for i in soup.findAll('div', class_='breed-type-card'):
        urls.append(i.a.get('href'))
   
 ### find all the div tags with class and extract href for next page untill condition become flase ######

    try:
        if soup.find('div', class_='mt2 bpm-mt5').a.get('href'):
            dogs(soup.find('div', class_='mt2 bpm-mt5').a.get('href')) 
    except:
        print("No More Pages")
 


dogs("https://www.akc.org/dog-breeds/")        
       
a = {
       "Names":names,
       "Urls" : urls
    }

df=pd.DataFrame(a)
df.to_csv('dogsa.csv',index=False)

df
```
